```python
matrix = data.pivot_table(
    index='userId',
    columns='title',
    values='rating'
)
```

```python
matrix
```

```python
def common_ratings(mymovie):
    common_ratings_dict = dict()
    non_nulls_mymovie = (~pd.isnull(matrix[mymovie]))
    for movie in matrix:
        non_nulls_movie = (~pd.isnull(matrix[movie]))
        common_ratings_dict[movie] = ((non_nulls_mymovie == non_nulls_movie) & (non_nulls_movie != False)).sum()
    return common_ratings_dict
```

```python
def n_similar_movies(movie_title, n_recommendations, common_weight = 0, min_common = 0):
    corrs = matrix.corrwith(matrix[movie_title])
    common_ratings_pd = pd.Series(common_ratings(movie_title))
    sim_movies = pd.concat([corrs.rename('correl'), common_ratings_pd.rename('common')], axis = 1)
    sim_movies.drop(movie_title, axis = 0, inplace = True)
    sim_movies = sim_movies.loc[sim_movies['common'] >= min_common]
    sim_movies['score'] = (1 - common_weight) * sim_movies['correl'] + common_weight * sim_movies['common']
    return sim_movies.nlargest(n_recommendations, 'score')           
```

```python
print(n_similar_movies('Toy Story (1995)', 10, 0.001, 10))
```

```python
print(n_similar_movies('Forrest Gump (1994)', 10, 0.001, 10))
```
